```python
from stdmodandoption import *
import plot_setup as PS
import collections
from CR_z_energy_distribution import CR_z_energy_distribution

def main():
    startno=490
    Nsnap=500
    snapsep=10
    wanted='dummy'
    dirdict = collections.defaultdict(dict)
    dirdict['plot1']=['bwmwmrdc27','bwmwmrdc28','bwmwmrdc29']
    noplots=len(dirdict.keys())
    fig, ax = PS.setupfig(nrows=1, ncols=noplots,sharex=True,sharey=True)
    fmeat=''
    noitem = len(dirdict.items())
    for i, (key,items) in enumerate(dirdict.items()):
        print 'i,key', i,key
        for runtodo in items:
            print 'runtodo', runtodo
            plotdict = CR_z_energy_distribution(runtodo,wanted,startno,Nsnap,snapsep,fmeat)
            xnl = plotdict[wanted]['xnl'];
            ynl = plotdict[wanted]['ynl'];
            xlab = plotdict[wanted]['xlab'];
            ylab = plotdict[wanted]['ylab'];
            labelneed = plotdict[wanted]['labelneed'];
            color = plotdict[wanted]['color'];
            lsn = plotdict[wanted]['lsn'];
            filename = plotdict[wanted]['filename'];
            ptitle = plotdict[wanted]['ptitle']
            if noitem==1:
                ax.plot(xnl,ynl,label=labelneed,lw=2,ls=lsn,color=color)
            else:
                ax[i].plot(xnl,ynl,label=labelneed,lw=2,ls=lsn,color=color)
        if np.remainder(i,noplots)>0: ylab=''
        if noitem==1:
            PS.miscsetup(ax,logx=0,logy=1,xlab=xlab,ylab=ylab,title=ptitle,legendneed=1,labfs=22,legfs=12)  
        else:
            PS.miscsetup(ax[i],logx=0,logy=1,xlab=xlab,ylab=ylab,title=ptitle,legendneed=1,labfs=22,legfs=12)  
    PS.finishsave(plt,filename)


        
if __name__ == '__main__':
    main()
```
