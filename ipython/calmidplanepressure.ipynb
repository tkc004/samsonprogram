```python
from samson_functions import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm



runtodo = 'm12icr_700hr'
Nsnap=600

rmax=20.; zmax=5.; nogrid=30

rneed = 0.0 #kpc

zlist = np.linspace(0.01,zmax,num=10)


G=readsnapfromrun(runtodo,Nsnap,0,rotface=1,loccen=1)

fitGdata = fitMNpot(G, rmax=rmax,zmax=zmax,nogrid=nogrid)

aG = fitGdata['afit']; bG = fitGdata['bfit']; MG = fitGdata['Mfit'];

rhoreal = fitGdata['rhoreal']; rlistr = fitGdata['rlist']; zlistr = fitGdata['zlist'];

rhoreal = rhoreal.reshape((len(rlistr),len(zlistr)))
                          
rhoatrl = np.array([])

for irho in range(len(zlistr)):
    rhoatr = np.interp(rneed, rlistr,rhoreal[:,irho])
    rhoatrl = np.append(rhoatrl,rhoatr)

rhorzl = np.interp(zlist, zlistr,rhoatrl)

print 'rhorzl', rhorzl

Gdata = calMNpot(rneed,zlist,MG,aG,bG)

GdPhidz = Gdata['dPhidz']

S = readsnapfromrun(runtodo,Nsnap,4,rotface=1,loccen=1)

fitSdata = fitMNpot(S, rmax=rmax,zmax=zmax,nogrid=nogrid)

aS = fitSdata['afit']; bS = fitSdata['bfit']; MS = fitSdata['Mfit'];

Sdata = calMNpot(rneed,zlist,MS,aS,bS)

SdPhidz = Sdata['dPhidz'] 

DM = readsnapfromrun(runtodo,Nsnap,1)

DMdata = DMtosphpot(DM,cutrout=20., nogrid=20.0)

DMdPhidr = DMdata['dPhidr']; rsph = DMdata['rsph']

rsphneed = np.sqrt(rneed*rneed+zlist*zlist)

DMdPhidrneed = np.interp(rsphneed,rsph,DMdPhidr)

DMdPhidz = DMdPhidrneed/rsphneed*zlist

rhog = rhorzl*(SdPhidz+GdPhidz+DMdPhidz)


#print 'DMdPhidz', DMdPhidz

#print 'MG, MS, MDM', MG, MS, DMdata['Menc'][-1]

plt.plot(zlist, SdPhidz)
plt.plot(zlist, GdPhidz,ls='dashed')
plt.plot(zlist, DMdPhidz, ls='dotted')






print 'rhog', rhog


plt.gcf().canvas.draw()
plt.show()

```
